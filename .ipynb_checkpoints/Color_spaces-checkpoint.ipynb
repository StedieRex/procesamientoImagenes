## Espacio de color RGB

El espacio de color RGB es el espacio de color m√°s difundido al momento de trabajar con imagenes digitales a color. Cada canal es una imagen a escala de gris, pero con diferente informaci√≥n de pixel.

```python
(b, g, r) = cv2.split(img)
```

Se muestran cada uno de los canales con MatPlotLib

```python
plt.figure(figsize=(15,10))
plt.subplot(131)
plt.imshow(r,cmap='gray')
plt.title('R Channel')
plt.subplot(132)
plt.imshow(g,cmap='gray')
plt.title('G Channel')
plt.subplot(133)
plt.imshow(b,cmap='gray')
plt.title('B Channel')
plt.show()
```

Mostrando las imagenes polarizadas

```python
import numpy as np
```

```python
zeros = np.zeros(img.shape[:2], dtype = "uint8")
```

```python
img_r = cv2.merge([r,zeros,zeros])
img_g = cv2.merge([zeros,g,zeros])
img_b = cv2.merge([zeros,zeros,b])

plt.figure(figsize=(15,10))
plt.subplot(131)
plt.imshow(img_r,cmap='gray')
plt.title('R Channel')
plt.subplot(132)
plt.imshow(img_g,cmap='gray')
plt.title('G Channel')
plt.subplot(133)
plt.imshow(img_b,cmap='gray')
plt.title('B Channel')
plt.show()
```

## Transformaci√≥n a Escala de Grises

Para transformar una imagen a escala de grisis, se puede optar por usar el m√©todo "cvtColor" de OpenCV o bien, implementar a trav√©s de f√≥rmulas la transformaci√≥n.

A trav√©s del m√©todo "cvtColor" de openCV, se puede transformar una imagen a color a escala de grises usando el par√°metro "COLOR_BGR2GRAY".

```python
# Usando la transformaci√≥n RGB -> GRAY de OpenCV (). Para las √∫ltmias versiones de OpenCV
# se usa la transformaci√≥n poderada (0.299, 0.587, 0.114) para los canales RGB respectivamente
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img_gray,cmap='gray')
```

Se puede optar por transformar la imagen a partir de una operaci√≥n ponderada. Por ejemplo:

```python
# Transformacion usando la ponderaci√≥n (0.299, 0.587, 0.114) para los canales RGB
w_r = 0.299
w_g = 0.587
w_b = 0.114
```

```python
gray_lum = w_r * r + w_g * g + w_b * b
```

```python
plt.imshow(gray_lum,cmap='gray')
```

```python
dim_gray = img_gray.shape
dim_lum = gray_lum.shape
print(dim_gray)
print(dim_lum)
```

```python
plt.figure(figsize=(15,10))
plt.subplot(121)
plt.imshow(img_gray,cmap='gray')
plt.title('Imagen Esc. Grises Promedio')

plt.subplot(122)
plt.imshow(gray_lum,cmap='gray')
plt.title('Imagen Esc. Grises Ponderado')
plt.show()
```

Observemos que la imagen generada a trav√©s de la f√≥rmula propuesta genera una matriz con valores flotantes

```python
# imprimiendo un pixel de la imagen generada
print(gray_lum[0][0])
```

Para evitar esto, se debe de crear un arreglo de tipo uint8

```python
gray_lum = np.array(gray_lum, dtype='uint8')
```

```python
# imprimiendo un pixel de la imagen generada
print(gray_lum[0][0])
```

Generando otras transformaciones a escala de grisis a partir de diferentes est√°ndares: ITU-BT.709, distribuci√≥n Gamma

```python
w_r = 0.2126
w_g = 0.7152
w_b = 0.0722
gray_TB_709 = np.array(w_r * r + w_g * g + w_b * b,cmap='uint8')
w_r = 0.309
w_g = 0.609
w_b = 0.082
gray_gamma = np.array(w_r * r + w_g * g + w_b * b)
```

```python
# imprimiendo las imagenes generadas previamente
plt.figure(figsize=(15,10))
plt.subplot(121)
plt.imshow(gray_TB_709,cmap='gray')
plt.title('Imagen ITU-BT.709')

plt.subplot(122)
plt.imshow(gray_gamma,cmap='gray')
plt.title('Imagen Esc. Grises Gamma')
plt.show()
```

## Transformaci√≥n RGB a HSV

Ejemplo para transformar im√°genes RGB/BGR a HSV

```python
# Dependiendo si la imagen de entrada es RGB o BGR, se usa la opci√≥n COLOR_RGB2HSV - COLOR_BGR2HSV
img_hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
(h, s, v) = cv2.split(img_hsv)
```

```python
#imprimiendo un pixel de cada uno de los canales
x = 0
y = 0
cadena = "H: p[{},{}] = {}".format(y,x,h[y,x])
print(cadena)
cadena = "S: p[{},{}] = {}".format(y,x,s[y,x])
print(cadena)
cadena = "V: p[{},{}] = {}".format(y,x,v[y,x])
print(cadena)
```

Nota: los valores generados al transformar un RGB a HSV son autom√°ticamente transformados por OpenCV a valores en el rango [0, 255] a trav√©s de las siguientes ecuaciones:

1. Canal H: h <- H / 2
2. Canal S: s <- S * 255
3. Canal V: p <- V * 255

Se debe considerar que en la transformaci√≥n, el resultado se guarda en un valor uint8.

```python
# imprimiendo los canales h, s, v
plt.figure(figsize=(15,10))
plt.subplot(131)
plt.imshow(h,cmap='gray')
plt.title('H Channel')

plt.subplot(132)
plt.imshow(s,cmap='gray')
plt.title('S Channel')

plt.subplot(133)
plt.imshow(v,cmap='gray')
plt.title('V Channel')
plt.show()
```

```python
# Prueba para comparar si la transformaci√≥n BGR TO HSV TO BGR genera valores iguales
img_hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
img_inv = cv2.cvtColor(img_hsv,cv2.COLOR_HSV2BGR)

# Compraci√≥n de las matrices
if np.array_equal(img,img_inv):
    print("Matrices iguales")
else:
    print("Matrices no iguales")
```

## Modelo CIE L*a*b

Otro modelo de especial inter√©s es CIE Lab. OpenCV ofrece a trav√©s de "cvtColor" la opci√≥n "cv2.BGR2LAB" o bien "cv2.RGB2LAB" para la transformaci√≥n a este espacio de color.

```python
#Lab 
lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
(l, a, b) = cv2.split(lab)
```

```python
# imprimiendo los canales l, a, b
plt.figure(figsize=(15,10))
plt.subplot(131)
plt.imshow(l,cmap='gray')
plt.title('L Channel')

plt.subplot(132)
plt.imshow(a,cmap='gray')
plt.title('a Channel')

plt.subplot(133)
plt.imshow(b,cmap='gray')
plt.title('b Channel')
plt.show()
```
